### Ensemble Learning

- Collection of Different Ml models combine together


- Ensemble learning follows concept of Wisdom of the crowd 



- Base Models should be different in 2 ways
    * Use diff ML algorithm 
    * Single Ml models but train all model with diff data.


- Types of Ensemble Learning


1) Voting Ensemble - Base models uses different ML algorithm.



2) Bagging - Base model use Single ML models but train all model with diff data.


3) Boosting - Weak learners are converted to strong learners,
                    previous models tells his error to next model and model gives more wieghtage to misclassified data.



4) Stacking - Different weights are given to diff algorithm based on o/p.



### When to Use - Always



### Benefits

1. Improvement in performance

2. Low Bias and Low variance can be acheived.

3. Robustness



### Disadvantages

- More Computation Power as we have to train multiple models.



### 1.Voting Ensemble 


### Assumptions

- All the Base ML models should be independent in nature

- All models should have minimum accuracy - 50 %

### Hard Voting vs Soft Voting 


- Hard Voting will work as normal voting .Check the o/p of all the model and take the mode .

### 2. Bagging

- Also known as Boostrap aggregation

- Multiple base models with same ML alogrithm

- Draw random samples from data for each model - bootstrap

- Model predict the o/p and bagging takes the majority count - aggregation.

### When to use bagging - Always


### Why use bagging

- To overcome the problem of Overfitting.
- If we have Low bias and High Variance.


### Bagging VS Random Forests

- Bagging can use different Base ML model such as DT , KNN , SVM.

- Random Forests uses base ML Models as Decision trees only.

- If we use bagging and use base model as decision tree then also it is not random forests.

- In Bagging if we use decision tree as base model , column sampling will be applied to entire tree , whereas the column sampling in random forest is applied for every nodes.

### 3.   Boosting


- Weak learners are converted to strong learners, previous models tells his error to next model and model gives more wieghtage to misclassified data.


###  4.    Stacking - Same as Voting Classifier


- Diff types of base ML model can be used

- Step 1 - Pass the dataset to 3 diff ML Base models

- Step 2 - Dataset : i/p = cgpa , iq , o/p = package

- Step 3 - Pass the CGPA,IQ input feature to ML base models.

- Step 4 - Every model will give an output , pass all 3 o/p to new model also known as meta models and check the prediction done by the final model.

- Combining o/p of first 3 model and pass as input to final model.

- First 3 model = LR + DT + KNN

- Final Model = RF

- DataSet is given to base model and prediction done by base model is given to meta model as input to train and predict the output.
